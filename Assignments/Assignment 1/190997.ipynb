## Q1. (a)
Train a k-nearest neighbors regression model, and report its validation set performance using root mean squared error.

## The model

```python
def kNN(x):
    """Predicts the target variable for x using kNN regression for a given k."""
    # calculating the differences for the numerical variables except the target variable
    diff = X_train[cols_numerical[:-1]] - x[cols_numerical[:-1]]
    for col in cols_nominal:
        # calculating the dissimilarity for nominal variables and concatenating it to diff
        diff_col = (X_train[col] != x[col])
        diff = pd.concat([diff, diff_col], axis=1)
    diff = diff.astype(float)
    # calculating the Euclidean distance
    dist = lin.norm(diff, 2, axis=1)
    idx = np.argsort(dist)
    y_pred = 0
    # calculating the average of the target variable of the k nearest neighbours of x
    for i in range(k):
        y_pred += y_train[idx[i]]
    return y_pred / k
```

```python
def kNN_predict():
    """Performs kNN regression on X_test and returns the predicted values."""
    return X_test.apply(kNN, axis=1)
```

```python
def get_rmse(y_pred):
    """Calculates the root mean squared error of y_pred."""
    diff = y_test - y_pred
    return lin.norm(diff, 2)
```

## Q2. (a)
Train a decision tree classification model using information gain as the splitting criterion and using only single feature decision stumps at all non-leaf nodes and majority votes at leaf nodes, and report its validation set performance using % accuracy. 

## The model

```python
class Node:
    # A node stores its own depth (root = depth 0), its decision stump, its parent and child information
    # Leaf nodes also store a constant label that is assigned to every data point that reaches that leaf
    def __init__(self, depth = 0, stump = (0,0), parent = None):
        self.depth = depth
        self.stump = stump
        self.parent = parent
        self.left = None
        self.right = None
        self.isLeaf = True
        self.label = 0
        
    def predict(self, data):
        # If it is a leaf node, it can predict rightaway
        if self.isLeaf:
            return self.label
        # Else, one of its children has to do the job
        else:
            if data[self.stump[0]] > self.stump[1]:
                return self.right.predict(data)
            else:
                return self.left.predict(data)
    
    # Gets the entropy of a node with nPos positive points and nNeg negative points
    def getEntropy(self, nPos, nNeg):
        if nPos == 0 or nNeg == 0:
            return 0
        nTot = nPos + nNeg
        # Finding the proportion of the positives and negatives in that node
        pPos = nPos / nTot
        pNeg = nNeg / nTot
        return - pPos * float(np.log(pPos)) - pNeg * float(np.log(pNeg))
            
    def getStump(self, X, y):
        n = y.size
        # Number of positive and negative labels at this node
        nPos = (y == 1).sum()
        nNeg = (y == -1).sum()
        bestObjective = -float('inf')
        
        # For each of the features in the data
        for i in range(X.shape[1]):
            # Do not use the same feature as used by the parent node
            if self.parent is not None and i == self.parent.stump[0]:
                continue
            # Finding out all values at which we can threshold that feature
            candidateThresholds = np.sort(X[:, i])
            idx = np.argsort(X[:, i])
            # The cumulative sum trick as the labels are binary 
            ySorted = y[idx]
            yCum = np.cumsum(ySorted)
            yCumRev = np.cumsum(ySorted[::-1])[::-1]
            # For each possible threshold (except the ones at the extreme)
            for j in range(1, candidateThresholds.size-1):
                # Calculating the information gain
                candidateObjective = self.getEntropy(nPos, nNeg) \
                                    - ((j + 1) / n) * self.getEntropy((yCum[j] + j + 1) / 2, (j + 1 - yCum[j]) / 2) \
                                    - ((n - j - 1) / n) * self.getEntropy((yCumRev[j+1] + n - j - 1) / 2, (n - j - 1 - yCumRev[j+1]) / 2)
                if candidateObjective > bestObjective:
                    bestObjective = candidateObjective
                    bestFeat = i
                    bestThresh = candidateThresholds[j]
        return bestFeat, bestThresh
            
    def train(self, X, y, maxLeafSize, maxDepth):
        # If too few data points are present, or else if this node is too deep in the tree, make this a leaf
        if y.size < maxLeafSize or self.depth >= maxDepth:
            self.isLeaf = True
            self.label = 1 if y.sum() >= 0 else -1
        else:
            # This node will be split and hence it is not a leaf
            self.isLeaf = False
            # Getting the best possible decision stump
            self.stump = self.getStump(X, y)
            self.left = Node(depth = self.depth + 1, parent = self)
            self.right = Node(depth = self.depth + 1, parent = self)
            # Finding which points go to the left child and which go to the right child
            discriminant = X[:, self.stump[0]] - self.stump[1]
            # Training the two children recursively
            self.left.train(X[discriminant <= 0, :], y[discriminant <= 0], maxLeafSize, maxDepth)
            self.right.train(X[discriminant > 0, :], y[discriminant > 0], maxLeafSize, maxDepth)
```

```python
class Tree:
    def __init__(self, maxLeafSize = 10, maxDepth = 5):
        self.root = Node()
        self.maxLeafSize = maxLeafSize
        self.maxDepth = maxDepth 
        
    def predict(self, x):
        return self.root.predict(x)
    
    def train(self, X, y):
        self.root.train(X, y, self.maxLeafSize, self.maxDepth)
```

```python
def dtree_predict(maxLeafSize, maxDepth):
    """Performs Decision Tree classification on X_test and returns the predicted labels."""
    tree = Tree(maxLeafSize, maxDepth)
    tree.train(np.array(X_train), np.array(y_train))
    return X_test.apply(tree.predict, axis=1)
```

```python
def get_accuracy(y_pred):
    """Calculates the accuracy of y_pred."""
    y_pred = dtree_predict(maxLeafSize, maxDepth)
    return (y_pred == y_test).sum() / len(y_test) * 100
```

### Training the model

```python
# training a Decision Tree classifier for maxLeafSize = 10 and maxDepth = 5 and calculating its accuracy
maxLeafSize = 10
maxDepth = 5
y_pred = dtree_predict(maxLeafSize, maxDepth)
accuracy = get_accuracy(y_pred)
print(f'Accuracy for Decision Tree classifier with max leaf size = {maxLeafSize} and max depth = {maxDepth}: {round(accuracy, 2)}%')
```

The accuracy of the decision tree classification model on the validation set is 83.04%

Use cross-validation to optimize the tree hyperparameters.

### Optimizing the hyperparameters

```python
# performing cross validation to find optimal values for maxLeafSize and maxDepth
maxLeafSizeOpt = 10
maxDepthOpt = 5
max_accuracy = accuracy
for maxDepth in range(3, 7):
    for maxLeafSize in range(15, 0, -5):
        y_pred = dtree_predict(maxLeafSize, maxDepth)
        accuracy = get_accuracy(y_pred)
        print(f'For max leaf size = {maxLeafSize} and max depth = {maxDepth}, accuracy = {round(accuracy, 2)}%')
        if accuracy > max_accuracy:
            max_accuracy = accuracy
            maxLeafSizeOpt = maxLeafSize
            maxDepthOpt = maxDepth
print()
print(f'Optimal value of maxLeafSize: {maxLeafSizeOpt}')
print(f'Optimal value of maxDepth: {maxDepthOpt}')
```
